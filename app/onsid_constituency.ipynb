```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
# The following set of code will scrape the Constituency Names for Northern Ireland and create a Dataframe with both the ONS_id
# and the constituency name.
emptydf = pd.DataFrame(columns=['ons_id', 'constituency_name'])
base_url = 'https://bbc.com/news/politics/constituencies/N060000'
ons_id_base='N060000'
for i in range (1,19):
    if i<10:
        url=base_url+'0'+str(i)
        ons_id=ons_id_base+'0'+str(i)
        print(ons_id)
    elif i>=10:
        url=base_url+str(i)
        ons_id=ons_id_base+str(i)
        print(ons_id)
    response=requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    results=soup.find('div', class_="constituency-title").h1.text
    emptydf.loc[i]=[ons_id, results]
    print(i, results)
emptydf
    
```
